# **DBSCAN**
**Density-based spatial clustering of applications with noise**

## Imports & Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
nokta_df = pd.read_csv("nokta.csv")
hilal_df = pd.read_csv("hilal.csv")
halka_df = pd.read_csv("halka.csv")
```

---
## DBSCAN Uygulama

```python
def kume(model, data):
    label = model.fit_predict(data)

    fig, axes = plt.subplots(2,1, figsize=(5,10), dpi=80)

    sns.scatterplot(ax=axes[0], x="p1", y="p2", data=data)    
    sns.scatterplot(ax=axes[1], x="p1", y="p2", data=data, hue=label, palette="viridis")    
```

### K - Means

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=2)

kume(model,nokta_df)
```

```python
model = KMeans(n_clusters=2)

kume(model,hilal_df)
```

```python
model = KMeans(n_clusters=2)

kume(model,halka_df)
```

### DBSCAN 

```python
from sklearn.cluster import DBSCAN
model = DBSCAN(eps=2)
kume(model,nokta_df)
```

```python
model = DBSCAN(eps=0.2)
kume(model,hilal_df)
```

```python
model = DBSCAN(eps=0.2)
kume(model,halka_df)
```
