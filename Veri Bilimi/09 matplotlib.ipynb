## Grafik Düzenleme

```python
x=np.arange(0,10)
```

```python
y=x**2 -10*x+25
```

```python
x,y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(12,6),dpi=200)  #boyut ve kalite
plt.plot(x,y)
plt.title("Başlık",size=18)
plt.xlabel("X ekseni",size=15)
plt.ylabel("(x-5)^2",size=15)
plt.xticks(np.arange(min(x)-1,max(x)+1,1));
plt.yticks(np.arange(min(y)-2.5,max(y)+1,2.5));
#plt.xlim(3,7)
# plt.ylim(0,5)
plt.show()
```

## Figür Komutları

```python
fig=plt.Figure()
axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.2,0.2,0.5,0.5])

axes1.plot(x,y)
axes1.set_xlabel("X Ekseni")
axes1.set_ylabel("y Ekseni")
axes1.set_title("Büyük Başlık")

axes2.plot(x,y)

axes2.set_title("Küçük Başlık")
```

```python
fig=plt.Figure()

axes1=fig.add_axes([0,0,1,1])
axes2=fig.add_axes([0.7,0.7,0.2,0.2])

axes1.plot(x,y)
axes2.plot(x,y)
```
