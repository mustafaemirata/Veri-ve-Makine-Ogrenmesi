## GridSearch

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("diyabet.xlsx")
df.head()
```

```python
df.drop("Y",axis=1,inplace=True)

X=df.drop("S6",axis=1)
y=df["S6"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=9)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
scaler.fit(X_train)

X_train=scaler.transform(X_train)
X_test=scaler.transform(X_test)
```

## Grid Serch

1. Tahmin edilecek algoritmayı kullan
2. parametreleri belirle.
3. verileri grid search' oturt.
4. cv unutma!!
5. puan fonksiyonu belirle

```python
from sklearn.linear_model import ElasticNet
```

```python
elasticnet=ElasticNet()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
params={
    "alpha":[.01,.1,1,5,10,50,100,1000],
    "l1_ratio":[.01,0.01,0.1,.5,.7,.9,.95,.99]
}
```

```python
grid_model=GridSearchCV(
    estimator=elasticnet, #tahmin edilecek
    param_grid=params,
    scoring="neg_mean_squared_error",
    cv=5, #5 eşit parça için.
    verbose=2
)
```

```python
grid_model.fit(X_train,y_train)
```

```python
grid_model.best_estimator_ #en iyi sonucu aldığımız değerler:
```
