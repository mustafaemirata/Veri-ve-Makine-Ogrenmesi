```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv("boston.csv")
```

## Train Test Split


```python
X=df.drop("PRICE",axis=1)
y=df["PRICE"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=9)
```

```python
df.shape
```

```python
X_train.shape #bu veriyle öğrenecek diğeriyle test edilecek.
```

```python
X_test.shape
```

```python
X_test.head()
```

## Regr Model

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_regr=LinearRegression()
```

```python
lin_regr.fit(X_train,y_train)
```

```python
lin_test_pred=lin_regr.predict(X_test)
```

```python
lin_test_pred
```

## Metrikler

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae=round(mean_absolute_error(y_test,lin_test_pred),2)
rmse=round(np.sqrt(mean_squared_error(y_test,lin_test_pred)),2)
r2=round(r2_score(y_test,lin_test_pred),2)

print("MAE: ",mae)
print("RMSE: ",rmse)
print("R2: ",r2)
```

```python
print(F"Ortalama Fiyat: {round(df['PRICE'].mean())}")
print(F"Standart Sapma: {round(df['PRICE'].std(),2)}")
```

# Poly Regr

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_conv=PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly_feat=poly_conv.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    poly_feat, y, test_size=0.2, random_state=9)
```

```python
poly_model=LinearRegression(fit_intercept=True)
```

```python
poly_model.fit(X_train,y_train)
```

```python
poly_test_pred=poly_model.predict(X_test)
```

```python
mae=round(mean_absolute_error(y_test,poly_test_pred),2)
rmse=round(np.sqrt(mean_squared_error(y_test,poly_test_pred)),2)
r2=round(r2_score(y_test,poly_test_pred),2)

print("MAE: ",mae)
print("RMSE: ",rmse)
print("R2: ",r2)

print(F"Ortalama Fiyat: {round(df['PRICE'].mean())}")
print(F"Standart Sapma: {round(df['PRICE'].std(),2)}")
```
