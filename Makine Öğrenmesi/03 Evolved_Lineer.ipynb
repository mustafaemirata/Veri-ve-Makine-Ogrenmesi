## Lineer Regresyon

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("community_health_MLR.csv")
```

```python
df.head()
```

## Veri DÃ¼zenleme

```python
df["vki"]=round(df["kilo"]/(df["boy"]/100)**2,2)
df.head()
```

```python
df.to_csv("community_health_evolved.csv",index=False)
```

```python
sns.scatterplot(data=df, x="visits",y="vki")
```

```python
sns.pairplot(df)
```

```python
df.corr(numeric_only=True)
```

## Train Test Split

```python
X=df.drop("visits",axis=1)
y=df["visits"]

#axis=0 satÄ±r
#axis=1 sÃ¼tun 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X.head()
```

```python
y.head()
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,
    test_size=0.33, random_state=99)
```

```python
X_train.head()
```

## Lin regr

```python
from sklearn.linear_model import LinearRegression
```

```python
regr=LinearRegression()
```

```python
regr.fit(X_train, y_train)
```

```python
test_pred=regr.predict(X_test)
```

```python
test_pred
```

## Performans

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
```

```python
print("Theta: ",round(regr.coef_[0],2))
print("intercept: ", round(regr.intercept_,2))
print("R^2: ",round(r2_score(y_test, test_pred),2))
print("MAE: ",round(mean_absolute_error(y_test,test_pred),2))
print("RMSE: ",round(np.sqrt(mean_squared_error(y_test,test_pred)),))
```

## Final

```python
final=LinearRegression()
```

```python
final.fit(X,y)
```

```python
y_hat=final.predict(X)
```
